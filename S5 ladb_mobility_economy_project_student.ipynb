### 2.3 Corregir formatos numéricos y de fecha

**🎯Objetivo:**
Asegurar que las columnas de fechas y valores numéricos estén en formatos correctos para permitir análisis, cálculos y comparaciones precisas.

**Instrucciones:**

- Convierte las columnas de fecha de `traffic` a formato `datetime`. Haz el cambio a prueba de errores.
- En el dataset `eco`, limpia los valores numéricos:
- En `city_gdp_capita`: elimina separadores de miles (`.`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- En `unemployment_pct`: elimina el símbolo de porcentaje (`%`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- En `population_m`: reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- Finalmente, crea una nueva columna llamada `population` multiplicando `population_m` por 1,000,000 para obtener la población total.


<details>
<summary>Haz clic para ver la pista</summary>
para eliminar símbolos, puedes reemplazarlos por un texto vacío.

```python
# Convertir las columnas de traffic a tipo fecha con pd.to_datetime()
traffic['update_time_utc'] = pd.to_datetime(traffic['update_time_utc'], errors='coerce', utc=True) #tu código aquí
traffic['update_time_utc_week_ago'] = pd.to_datetime(traffic['update_time_utc_week_ago'], errors='coerce', utc=True) #tu código aquí

# verificar el cambio
traffic.info()
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor          </b> <a class="tocSkip"></a>


Correcto, bien con el uso de `pd.to_datetime` para realizar la conversión a datetime de las columnas que contienen fechas. Bien al agregar `errors='coerce'` como parámetro, si hay una fila que no está en formato fecha, quedará como nulo.


</div>

```python
# Limpia separadores y convierte columnas numéricas en eco
eco['city_gdp_capita'] = eco['city_gdp_capita'].astype(str).str.strip().str.replace('.', '').str.replace(',', '.').astype(float)
eco['unemployment_pct'] = eco['unemployment_pct'].astype(str).str.strip().str.replace('%', '').str.replace(',', '.').astype(float)
eco['population'] = eco['population'].astype(str).str.strip().str.replace(',', '.').astype(float)

# Calcula la población total en unidades absolutas (Multiplica * 1000000)
eco['population'] = eco['population'] * 1000000 

# verificar el cambio
eco.info()
eco.head(3)
```

# 🧾 Resumen ejecutivo: Movilidad urbana y productividad económica.

**Contexto & objetivo:**  

La principales ciudades del mundo, enfrentan hoy en día un gran reto de movilidad urbana y su relación con la productividad económica. Los niveles de alta congestión son un factor que afectan los tiempos de viaje y generan retrasos, lo cual puede tener un efecto negativo a la productividad económica.

El presente análisis busca definir la relación existente entre los dos factores, por un lado el factor de alta congestión y por otro el PIB, como determinantes para una baja productividad económica. 

Para el desarrollo de este analisis, se utilizaron dos fuentes reales de datos: el de Movilidad Urbana de TomTom Traffic Index y el de Economía Urbana de OECD Cities. Entre las variables utilizadas para el analisis, se utilizaron los registros del 2024 de retrasos totales provocados por congestionamientos en las principales ciudades de latinoamerica, asi como el PIB per cápita para conocer el nivel de productividad económica por habitante en las mismas ciudades. 

**Cobertura de datos:**  

Los datos analizados del año 2024 fueron:

  - Ciudades: Belo Horizonte, Bogotá, Brasilia, Buenos Aires, Curitiva, Fortaleza, Lima, Ciudad de México, Montevideo, Porto Alegre, Recife, Rio de Janeiro. Salvador, Santiago, Sao Paolo.
  - Paises: Brasil, Colombia, Argentina, Perú, México, Uruguay y Chile. 

**Metodología (alto nivel):**  

Para el analisis se llevaron a cabo los sigiente procesos:

  - Carga y exploración de datos.
  - Limpieza y correción de formatos.
  - Extracción y filtrado del año 2024.
  - Calculo de estadisticas.
  - Combinación de datasets: tráfico y economía.
  - Visualización de relación entre variables.

Para llevar a cabo la integración de los dos Datasets a una sola base de datos, se hizo uso de la unión INNER, la cual se formulo con los llaves comunes ciudad-año.

Las validaciones visuales utilizadas para este analisis fueron, un boxplot para visualizar la distrubución del trafico en relación a su media, mediana y valores atipicos. Un histograma para visualizar la distribución de la economía, su forma y valor promedio. Tambien se utilizó una gráfica de barras para visualizar la relación entre congestión y PIB poblacional.

**Hallazgos iniciales:**  

De acuerdo con las analisis realizados, podemos definir que no existe una relación clara, por la que podamos definir que a mayor PIB per cápita se genere una mayor congestión.

Si bien para el caso de la CDMX y Sao Paulo, existe relación de PIB poblacional con congestión, tenemos el caso contrario para las ciudades de Brasilia y Montevideo. Por lo que pudieramos atribuir el fenomeno de congestionamiento a un tema de infraestructura y politicas publicas de las ciudades.

Existen valores anomalos del trafico que se pueden visualizar en el Boxplot, outliers que se ven muy por fuera de nuestra caja de valores. Es necesario llevar a cabo una revision adicional para determinar si estos valores son reales o se deben a algun otro factor.

- Realizar un análisis adicional y mas a detalle en la ciudad de Santiago, la cual es en general la ciudad de mayor congestión en relación a su PIB. Es necesario programas de inversion para mejoramiento urbano y eficiencia de transportes públicos.

- Para el caso de las ciudades de Brasilia y Montevideo son modelos de análisis, al ser las ciudades mejor evaluadas en relación a su correlación entre los valores de congestión vehicular y altos indicadores de eficiencia económica.

- En las Ciudad de México y Sao Paulo hay oportunidades para realizar inversiones en infraestructura de transporte, que mejoren la movilidad urbana e incrementen la productividad económica.
